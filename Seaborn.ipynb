# Join Plot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

#univariate KDE plot

> vishulaizing the distribution of a single variable (eg: 'sepal length(cm)')



```python
df = sns.load_dataset('iris')

plt.figure(figsize=(8, 6))
sns.kdeplot(data = df, x = 'sepal_length', fill = True, alpha = 0.5,)
plt.title('univariate KDE plot of sepal length')
plt.xlabel('sepal length (cm)')
plt.ylabel('density')
plt.show()
```

```python
df = sns.load_dataset('iris')

plt.figure(figsize=(8, 6))
sns.kdeplot(data = df, x = 'sepal_length', fill = True, alpha = 0.5, hue = 'species')
plt.title('univariate KDE plot of sepal length')
plt.xlabel('sepal length (cm)')
plt.ylabel('density')
plt.show()
```

#bivariate KDE plot

> vishualizing the join distibution of tow variable(eg: "sepal length" (cm) and "sepal width" (cm))



```python
plt.figure(figsize = (8, 6))
sns.kdeplot(data = df, x = "sepal_length", y = "sepal_width", fill = True, alpha = 0.5, hue = "species")

plt.title("BIVARIATE KDE PLOT")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```
