#annottation with arrow

```python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Create a line chart
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')
# Add a basic annotation with an arrow
plt.annotate( 'Annotation with arrow',
            xy=(3, 6),
            xytext=(4, 7),
            arrowprops=dict(facecolor='black', arrowstyle='->')
)
# Add title and labels
plt.title('Basic Annotation with Arrow')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
# Display grid
plt.grid(True)
# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Create a line chart
plt.figure (figsize=(8, 6))
plt. plot(x, y, marker='o', linestyle='-')
# Add multiple annotations with different styles
annotations = [
              {'xy': (1, 2), 'text': 'Point 1', 'xytext': (2, 3), 'arrowprops': dict(facecolor='blue', arrowstyle='->')},
              {'xy': (3, 6),'text': 'Point 2','xytext': (4, 5), 'arrowprops': dict(facecolor='green', arrowstyle='->')},
              {'xy': (5, 10),'text': 'Point 3', 'xytext': (4, 11), 'arrowprops': dict(facecolor='red', arrowstyle='->')}
]
for ann in annotations:
    plt.annotate(
            ann['text'],
          xy=ann['xy'],
          xytext=ann['xytext'],
          arrowprops=ann['arrowprops']
    )
plt.title('Multiple Annotations with Different Styles')
plt.xlabel('X-axis Label')
plt.ylabel ('Y-axis Label')
plt.grid(True)
plt.show()
```

##pie

```python
cars = ['alto','crysta','bmw','tesla','jaguer']
data = [23, 35, 29, 12, 33]

# customized option
explode = (0.1, 0, 0.2, 0.3, 0)

colors = ['orange', 'red', 'yellow', 'brown', 'indigo']

plt.pie(data, explode = explode, labels = cars, colors = colors,
        autopct='%.1f%%', shadow = True, startangle = 90)

plt.legend(cars, title = 'cars', loc= 'center left',
           bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('customizing pie chart')

plt.show()
```

#set border for wedges

```python
year = [2019,1022,2011,1999,1890]
profit = [15,19,35,14,17]

plt.pie(profit, labels = year,autopct='%1.1f%%',
        startangle = 50,
        wedgeprops = {'edgecolor' : 'black',
                      'linewidth' : 2,
                      'antialiased' : True })
plt.axis('equal')

plt.show()
```
